## CO2 emissions

Process file `CO2_emission.csv`. Build a model, that would predict CO2 emissions from the size of the engine of car. Visualize model on a plot.

Choose regression model that would best fit the data. Choose the library for model implementation (sklearn, other library, custom implementation).

```python
def CO2_emissions(file_path):
    
    data = pd.read_csv(file_path)
    
    # Select relevant columns for the model
    X = data[['Engine_Size']]
    y = data['CO2_Emissions']
    
    # Initialize and fit the linear regression model
    model = LinearRegression()
    model.fit(X, y)
    
    # Plot the data points and the regression line
    plt.scatter(X, y, color='blue', label='Actual Data')
    plt.plot(X, model.predict(X), color='red', label='Predicted Line')
    plt.xlabel('Engine Size')
    plt.ylabel('CO2 Emissions')
    plt.legend()
    plt.show()
    
    # Return model parameters (e.g., slope and intercept)
    return model.coef_.tolist(), model.intercept_


CO2_emissions('resources/CO2_emission.csv')
```

## Electric Power Consumption

Process file `powerconsumption.csv`. Build a model, that would predict electricity consumption from the weather: temperature, humidity and wind speed. Visualize model on a plot.

Choose regression model that would best fit the data. Choose the library for model implementation (sklearn, other library, custom implementation).

```python
def power_consumption(file_path):
    # Read the data
    data = pd.read_csv(file_path)
    
    # Use the correct column names
    X = data[['Temperature', 'Humidity', 'WindSpeed']]  # Note: WindSpeed not Wind_Speed
    y = data[['PowerConsumption_Zone1', 'PowerConsumption_Zone2', 'PowerConsumption_Zone3']].sum(axis=1)  # Sum the zones for total consumption
    
    # Initialize and fit the linear regression model
    model = LinearRegression()
    model.fit(X, y)
    
    # Make predictions
    predictions = model.predict(X)
    
    # Plot the actual vs predicted values
    plt.scatter(range(len(y)), y, color='blue', label='Actual')
    plt.plot(range(len(predictions)), predictions, color='orange', label='Prediction')
    plt.xlabel('Sample Index')
    plt.ylabel('Total Power Consumption')
    plt.legend()
    plt.show()
    
    # Return model coefficients
    return model.coef_.tolist()


file_path = 'resources/powerconsumption.csv'
power_consumption(file_path)
```

## Heart Conditions Predictions

Process file `heart.csv`. It has the following data:

- age - Age of the patient
- sex - Sex of the patient
- cp - Chest pain type ~ 0 = Typical Angina, 1 = Atypical Angina, 2 = Non-anginal Pain, 3 = Asymptomatic
- trtbps - Resting blood pressure (in mm Hg)
- chol - Cholestoral in mg/dl fetched via BMI sensor
- fbs - (fasting blood sugar > 120 mg/dl) ~ 1 = True, 0 = False
- restecg - Resting electrocardiographic results ~ 0 = Normal, 1 = ST-T wave normality, 2 = Left ventricular hypertrophy
- thalachh - Maximum heart rate achieved
- oldpeak - Previous peak
- slp - Slope
- caa - Number of major vessels
- thall - Thalium Stress Test result ~ (0,3)
- exng - Exercise induced angina ~ 1 = Yes, 0 = No
- output - Target variable - risk of heart attack - 1 is high chance of heart attack and 0 is low chance.

Use classification model (choose model best fitting model yourself). Choose parameters that would best fit for making a prediction on. Visualize model on a plot.

*Bonus points*: split data into train data and validation data. Validate your model.

```python
def heart_classification(file_path):

    data = pd.read_csv(file_path)
    
    # Preprocess the data: Convert categorical variables to numeric
    label_encoder = LabelEncoder()
    categorical_columns = ['sex', 'cp', 'restecg', 'fbs', 'exng', 'slp', 'caa', 'thall']
    for col in categorical_columns:
        data[col] = label_encoder.fit_transform(data[col])

    # Split the data into features (X) and target (y)
    X = data.drop(columns=['output'])
    y = data['output']
    
    # Standardize the features to improve convergence
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
    
    # Initialize and train the logistic regression model with 'saga' solver and higher iteration limit
    model = LogisticRegression(solver='saga', max_iter=5000, random_state=42)
    model.fit(X_train, y_train)
    
    # Make predictions on the test set
    y_pred = model.predict(X_test)
    
    # Evaluate the model's performance using classification report
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    disp.plot(cmap=plt.cm.Blues)
    plt.show()
    
    # Cross-validation for model validation
    cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='accuracy')
    print(f"Cross-Validation Accuracy: {cv_scores.mean():.2f}  {cv_scores.std():.2f}")
    
    # Return the model's coefficients
    return model.coef_.tolist()


file_path = 'resources/heart.csv'
heart_classification(file_path)
```

There are more files in the `resources` folder. Take a look at the. Visualize data in these files. Choose ML algorithm for building a model and making predictions for new data. Visualize the model.

The file telecom_cus.csv contains customer data for a telecom company. Here's a breakdown of the columns:
- region: Geographic region of the customer.
- tenure: Duration the customer has been with the company.
- age: Age of the customer.
- marital: Marital status of the customer.
- address: Type of address the customer has (e.g., urban, rural).
- income: Income level of the customer.
- ed: Education level of the customer.
- employ: Employment status of the customer.
- retire: Retirement status of the customer.
- gender: Gender of the customer.
- reside: Residential area (e.g., city or town).
- custcat: Customer category (the target variable, possibly with categories like 'high', 'medium', 'low' or binary classification).

```python
def telecom_classification(file_path):

    data = pd.read_csv(file_path)
    
    # Preprocess the data: Convert categorical variables to numeric
    label_encoder = LabelEncoder()
    
    categorical_columns = ['region', 'marital', 'address', 'ed', 'employ', 'retire', 'gender', 'reside']
    for col in categorical_columns:
        data[col] = label_encoder.fit_transform(data[col])
    
    # Split the data into features (X) and target (y)
    X = data.drop(columns=['custcat'])
    y = data['custcat']
    
    # Standardize continuous variables
    scaler = StandardScaler()
    X[['tenure', 'age', 'income']] = scaler.fit_transform(X[['tenure', 'age', 'income']])
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    # Initialize and train the model (Random Forest Classifier)
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Evaluate the model
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    
    # Confusion Matrix Visualization
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
    disp.plot(cmap=plt.cm.Blues)
    plt.show()

    # Return model's feature importance for insights
    feature_importance = model.feature_importances_
    feature_names = X.columns
    feature_df = pd.DataFrame({
        'Feature': feature_names,
        'Importance': feature_importance
    }).sort_values(by='Importance', ascending=False)
    
    print("\nFeature Importances:")
    print(feature_df)


file_path = 'resources/telecom_cus.csv'
telecom_classification(file_path)
```

- I used Random Forest Classifier, as it handles feature interactions well and gives us insights into feature importance.
- For performance evaluation and confusion matrix for visualization I used classification_report.
- The model also outputs the importance of each feature, which helps understand which features are most influential in predicting the target variable.
